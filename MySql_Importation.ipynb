```python
import pandas as pd
import mysql.connector

# Read the CSV file
file_path = "dataspark.csv"  # Your CSV file path
data = pd.read_csv(file_path)

# MySQL connection setup
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="1234",  # Replace with your MySQL password
    database="datapy"  # Replace with your database name
)
cursor = conn.cursor()

# Create a table (if it doesn't already exist)
table_name = "dataspark"  # Replace with your desired table name
create_table_query = f"""
CREATE TABLE IF NOT EXISTS {table_name} (
    order_number INT,
    line_item INT,
    order_date TEXT,  -- Store dates as TEXT
    customerkey INT,
    storekey INT,
    productkey INT,
    quantity INT,
    currency_code VARCHAR(10),
    gender VARCHAR(10),
    customers_city VARCHAR(50),
    customers_state VARCHAR(50),
    customers_country VARCHAR(50),
    customers_continent VARCHAR(50),
    age INT,
    product_name VARCHAR(255),
    brand VARCHAR(100),
    color VARCHAR(50),
    unit_cost_usd FLOAT,
    unit_price_usd FLOAT,
    subcategorykey INT,
    subcategory VARCHAR(100),
    categorykey INT,
    category VARCHAR(100),
    store_country VARCHAR(50),
    store_state VARCHAR(50),
    square_meters INT,
    open_date TEXT,  -- Store dates as TEXT
    exchange FLOAT,
    total_price FLOAT
)
"""
cursor.execute(create_table_query)

# Insert data into the table using parameterized queries
insert_query = f"""
INSERT INTO {table_name} (
    order_number, line_item, order_date, customerkey, storekey, productkey, quantity, 
    currency_code, gender, customers_city, customers_state, customers_country, 
    customers_continent, age, product_name, brand, color, unit_cost_usd, unit_price_usd, 
    subcategorykey, subcategory, categorykey, category, store_country, store_state, 
    square_meters, open_date, exchange, total_price
) VALUES (
    %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
)
"""

# Iterate through the DataFrame and insert rows safely
for _, row in data.iterrows():
    cursor.execute(insert_query, (
        row['order_number'], row['line_item'], str(row['order_date']),
        row['customerkey'], row['storekey'], row['productkey'], row['quantity'],
        row['currency_code'], row['gender'], row['customers_city'], row['customers_state'],
        row['customers_country'], row['customers_continent'], row['age'], 
        row['product_name'], row['brand'], row['color'], row['_unit_cost_usd_'], 
        row['_unit_price_usd_'], row['subcategorykey'], row['subcategory'], row['categorykey'], 
        row['category'], row['store_country'], row['store_state'], row['square_meters'], 
        str(row['open_date']), row['exchange'], row['total_price']
    ))

# Commit and close the connection
conn.commit()
cursor.close()
conn.close()

print("Data uploaded successfully!")
```
